```python
#4)4)	Consider the url ‘http://www2.sas.com/proceedings/sugi30/toc.html’.  Extract all the papers names, from 001-30 to 268-30.

from urllib.request import urlopen
from bs4 import BeautifulSoup

u = urlopen('http://www2.sas.com/proceedings/sugi30/toc.html')
b = BeautifulSoup(u, "html.parser")

p = b.findAll('cite')

for name in p:
    print(name.get_text(), sep = "\n")
```

```python
#5)	Consider the url=’http://www.dictionary.com/browse/’ and the words ‘handy’,’whisper’,’lovely’,’scrape’.Build a data frame, where the first variable is “Word” and the second variable is “definitions”. Scrape the definitions from the url.


import requests
from bs4 import BeautifulSoup
import pandas as pd

words = ['handy','whisper','lovely','scrape']

definitions = []

for x in words:
    
    req = requests.get("http://www.dictionary.com/browse/" )
    document = BeautifulSoup(req.content)
    
    finding  = document.find('span',{'class':'one-click-content css-1p89gle e1q3nk1v4'})
    
    con = finding.get_text(strip=True)
    definitions.append(con)
    
df = pd.DataFrame(list(zip(words,definitions)), columns = ['Word','Definition'])
df
```

```python
#7)	Select any popular ecommerce website such as AMAZON, FLIPCART, ALIBABA, DARAZ, etc. to do the following exercises.
#a.	Scrap the selected website and collect all the product related information from the website and store them in a data frame.


import pandas as pd
import re
from urllib.request import urlopen
from bs4 import BeautifulSoup
import requests

req = requests.get('https://www.amazon.in/gp/bestsellers/boost/ref=zg_bs_nav_10')
cont = req.content
soup_b = BeautifulSoup(cont)
 
p_name =[]; 
p_price =[];
p_review =[];
p_rate = [];
    
for d in soup_b.findAll('div', attrs={'class':'a-section a-spacing-none aok-relative'}):
    name = d.find('span', attrs={'class':'zg-text-center-align'})
    l = name.find_all('img', alt=True)
    ratings = d.find('span', attrs={'class':'a-icon-alt'}).text
    users_rated = d.find('a', attrs={'class':'a-size-small a-link-normal'}).text
    price = d.find('span', attrs={'class':'p13n-sc-price'}).text
    
    p_name.append(l[0]['alt'])
    p_price.append(price)
    p_review.append(ratings)
    p_rate.append(users_rated)
    
df = pd.DataFrame({'Product Name':p_name ,'Price':p_price, 'Rate':p_rate,'Review':p_review})
df




```

```python
#b.	Using the results from (a) identify and print/display the followings
#i.	Top 10 most popular products based on sales.
#ii.	Top 10 most popular products based on customer reviews.


data = df.sort_values(["Review"], axis=0, ascending=False)[:10]
data
```
